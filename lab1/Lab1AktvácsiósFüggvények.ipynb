```python
#1. küszöbfüggvény vagy egységsugarú függvény
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def thresholdfunction(t, theta):
    return (t>=theta) * 1

t = np.linspace(-5, 5)
theta = 0
plt.plot(t, thresholdfunction(t, theta))
plt.title('Threshold Activation Function')
plt.show()
```

```python
#2. Lépcsőfüggvény
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def stairfunction(t, theta):
    return (t<=theta) * -1 + (t>=theta) * 1

t = np.linspace(-5, 5)
theta = 0
plt.plot(t, stairfunction(t, theta))
plt.title('Stair Activation Function')
plt.show()
```

```python
#3. Lineáris aktívációs függvény
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def linearActivationFunction(t, theta):
    return t-theta

t = np.linspace(-5, 5)
theta = 1
plt.plot(t, linearActivationFunction(t, theta))
plt.title('Linear Activation Function')
plt.show()
```

```python
#4. Logisztikus vagy szigmoid alakú karakterisztika
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def SigmoidBinary(a, t, theta):
    return 1/(1+np.exp(-a*t-theta))

theta = 0
a = 1
t = np.linspace(-5, 5)
plt.plot(t, SigmoidBinary(a, t,theta))
plt.title('Binary Sigmoid Activation Function')
plt.show()
```

```python
#5. Tangens hiberbolikus függvény
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def tangensHyperbolic(a, t, theta):
    return (1-np.exp(-a*t-theta))/(1+np.exp(-a*t-theta))

theta = 0
a = 1

t = np.linspace(-5, 5)
plt.plot(t, tangensHyperbolic(a, t, theta))
plt.title('Tangens Hyperbolic Activation Function')
plt.show()
```

```python
#6. GAUSS
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def gaussSimple(t):
  return np.exp(-np.square(t))

t = np.linspace(-5, 5)
plt.plot(t, gaussSimple(t))
plt.title('Gauss Simple Activation Function')
plt.show()

def gauss(c, t, sigma):
  return np.exp((-np.square(t-c))/2*(sigma*sigma))

c = 1
sigma = 1
t = np.linspace(-5, 5)
plt.plot(t, gauss(c, t, sigma))
plt.title('Gauss Activation Function')
plt.show()
```

```python
#7. RELU
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def RectifiedLinearUnit(t):
  lst=[]
  for i in t:
    if i>=0:
      lst.append(i)
    else:
      lst.append(0)
  return lst

arr = np.linspace(-5, 5)
plt.plot(arr, RectifiedLinearUnit(arr))
plt.title('Rectified Linear Unit Activation Function')
plt.show()
```

```python
#8. Leaky RELU
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def LeakyRectifiedLinearUnit(arr):
    lst=[]
    for i in arr:
        if i>=0:
            lst.append(i)
        else:
            lst.append(0.01*i)
    return lst

arr = np.linspace(-5, 5)
plt.plot(arr, LeakyRectifiedLinearUnit(arr))
plt.title('Leaky ReLU Activation Function')
plt.show()
```

```python
#9. ELU (Exponential Linear Unit) függvény
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def elu(arr):
    lst=[]
    for i in arr:
        if i<0:
            lst.append(0.5*(np.exp(i) -1))
        else:
            lst.append(i)
    return lst

arr = np.linspace(-5, 5)
plt.plot(arr, elu(arr))
plt.title('ELU Activation Function')
plt.show()
```

```python
#10. Softmax függvény
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def softmax(t):
    return np.exp(t) / np.sum(np.exp(t))

t = np.linspace(-5, 5)
plt.plot(t, softmax(t))
plt.title('Softmax Activation Function')
plt.show()
```
